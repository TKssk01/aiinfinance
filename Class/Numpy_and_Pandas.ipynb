### NumPy Excercise

### Portfolio Returns

First, we are going to calculate the return of a given portfolio using NumPy.

Define a weight matrix, which represents the percentage of money you are going to invest in each stock, and also the investment amount.

````python
investment = 1000
weights = [0.4, 0.4, 0.1, 0.1]

# Finally, transform your weights to a numpy array
weights = np.array(weights)
````

Now, we need to calculate the growth rate of each asset.
Remember that the formula for getting the growth rate is:

* (current_value - past_value) / past_value

Use the following functions:

* `np.diff()`: Computes the difference between each element and its predecessor along the specified axis.
* `array[:, :-1]`: Slices the array to exclude the last column, aligning the lengths for element-wise division.

````python
# Calculate daily returns
daily_returns = np.diff(numpy_data) / numpy_data[:, :-1]
````

Now, given the growth rate of the assets, the weights, and the investment,
first calculate the overall return of the portfolio using the dot product:

* $(R = \text{weight} \cdot \text{assets_growth_rate})$

Then we need to calculate the return on investment:

* $(\text{cumulative growth factor})(\text{initial investment})$

Where:

* $\text{cumulative growth factor} = \prod_{i=1}^{t} (1 + r_i)$

And

* $r_i$: is the portfolio growth rate at time $i$

**Note:** use the `np.cumprod()` function

````python
# Calculate portfolio growth rate
portfolio_growth_rate = np.dot(weights, daily_returns)
````

````python
# Calculate cumulative growth factor
cumulative_growth_factor = np.cumprod(1 + portfolio_growth_rate)

# Calculate the portfolio value over time
portfolio = cumulative_growth_factor * investment
````

### Beta

Beta is a measure of a stock's (or portfolio's) sensitivity to movements in the overall market.
It indicates how much the asset's returns move relative to the market returns.

The formula to calculate beta is:
$\beta = \frac{Cov(R_i, R_m)}{\sigma_{m}^{2}}$

Where:
* $Cov(R_i, R_m)$: is the covariance between the asset returns and the market returns.
* $\sigma_{m}^{2}$: is the variance of the market returns.

**Uses:**

* **Market Risk Measurement:** Beta measures systematic risk, which is the risk inherent to the entire market or market segment.
* **CAPM (Capital Asset Pricing Model):** Beta is used in CAPM to determine the expected return of an asset based on its beta and the expected market return.

In this case, the market returns should be the safest asset, or the target return asset. For example, if other than oil and gold we had fintech or other assets, the target might be the Nasdaq since it is an important indicator of the overall market growth.

**Note:** Remember that the covariance matrix returns different variances and covariances. In order to get the covariance of $R_i$ with $R_m$, select the upper right value of the covariance matrix.

### Bollinger Bands

Bollinger Bands consist of a middle band (SMA) and two outer bands (standard deviations away from the SMA).

The Bollinger Bands consist of three components:

* **Middle Band**: This is usually a Simple Moving Average (SMA) of the price over a specified period. (Use rolling window)
* **Upper Band**: This is the SMA plus a specified number of standard deviations of the price.
* **Lower Band**: This is the SMA minus a specified number of standard deviations of the price.

Compute the Bollinger Bands for the Nasdaq ETF.


### Let's practice

````python
# Set the window and the number of standard deviations
window = 30 # Approximately one month of trading days
stds = 2 # Bollinger Bands typically use 2 standard deviations above and below the moving average
````

````python
# Calculate the Middle Band (Simple Moving Average)
data_frame['Middle Band'] = data_frame['Nasdaq ETF'].rolling(window=window).mean()
data_frame
````

````python
# Calculate the Standard Deviation
data_frame['Standard Deviation'] = data_frame['Nasdaq ETF'].rolling(window=window).std()
data_frame
````

````python
# Calculate the Upper and Lower Bands
data_frame['Upper Band'] = data_frame['Middle Band'] + (stds * data_frame['Standard Deviation'])
data_frame['Lower Band'] = data_frame['Middle Band'] - (stds * data_frame['Standard Deviation'])
data_frame
````

````python
# Plot the data - Long coding box
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(data_frame.index, data_frame['Nasdaq ETF'], label='Nasdaq ETF', color='blue')
plt.plot(data_frame.index, data_frame['Middle Band'], label='Middle Band (SMA)', color='red')
plt.plot(data_frame.index, data_frame['Upper Band'], label='Upper Band', color='green')
plt.plot(data_frame.index, data_frame['Lower Band'], label='Lower Band', color='orange')

# Fill the area between the upper and lower bands
plt.fill_between(data_frame.index, data_frame['Lower Band'], data_frame['Upper Band'], color='gray', alpha=0.2)

# Adding titles and labels
plt.title('Nasdaq ETF with Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
````
