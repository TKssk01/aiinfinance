```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import seaborn as sns  # seabornをインポートする
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
file_pep = './pep.us.csv'
data = pd.read_csv(file_pep)
# データの先頭行を確認する
print(data.head())
```

```python
# 日付をdatetime型に変換
data['Date'] = pd.to_datetime(data['Date'])
# 2000年以降のデータにフィルタリング
data_2000_onwards = data[data['Date'] >= '2000-01-01']
# 日付でデータをソート
data_2000_onwards = data_2000_onwards.sort_values('Date')
# ラインチャートを作成
plt.figure(figsize=(12, 6))
plt.plot(data_2000_onwards['Date'], data_2000_onwards['Close'], label='Close Price')
plt.title('Stock Prices of PEP (2000 onwards)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
